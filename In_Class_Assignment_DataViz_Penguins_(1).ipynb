## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.title('Bill Length vs Bill Depth')
```

Interpretations: Penguins of the same type generally shared characteristics of bill length and depth. Adelie penguins had shorter beaks with greater depth, Gentoo penguins with more length and less depth, and Chinstrap penguins having long and deep beaks. Overall, there was a negative correlation between length and depth.
